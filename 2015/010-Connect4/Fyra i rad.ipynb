```python
def ass(a, b):
    if a != b:
        print "Assert failed"
        print 'Är: ',a
        print 'Bör:',b
        assert a == b
```

```python
class Board(object):
    def __init__(self,moves = ''):
        self.moves = []
        self.board = ['', '', '', '', '', '', '']
        for move in moves:
            self.move(int(move))
    def move(self,m):
        self.board[m] += self.next_marker()
        self.moves.append(m)
    def undo(self):
        m = self.moves.pop()
        self.board[m] = self.board[m][:-1]
    def next_marker(self):
        return 'XO'[len(self.moves)%2]
    def last_marker(self):
        return 'OX'[len(self.moves)%2]
    def display_simple(self):
        return '\n:' +'\n:'.join(self.board) + '\n 012345\n'
    def display(self):
        return '\n:' +'\n:'.join(self.board) + '\n 012345\n'
    def check_column(self):
        m = self.moves[-1]
        marker = self.last_marker()
        return marker * 4 in self.board[m]
        
test_moves()    
test_marker()
test_board()
test_check()
test_display_simple()
test_display()
```

```python
def test_moves():
    b = Board()
    assert b.moves == []
    b.move(2)
    assert b.moves == [2]
    b.move(3)
    assert b.moves == [2, 3]
    b.undo()
    assert b.moves == [2]
    b.undo()
    assert b.moves == []
```

```python
def test_marker():
    b = Board()
    assert b.next_marker() == 'X'  # X börjar alltid
    b.move(2)
    assert b.last_marker() == 'X'
    assert b.next_marker() == 'O'
    b.move(3)
    assert b.last_marker() == 'O'
    assert b.next_marker() == 'X'
    b.undo()
    assert b.last_marker() == 'X'
    assert b.next_marker() == 'O'
    b.undo()
    assert b.last_marker() == 'O'
    assert b.next_marker() == 'X'
```

```python
def test_board():
    b = Board()
    assert b.board == ['', '', '', '', '', '', '']
    b.move(2)
    assert b.board == ['', '', 'X', '', '', '', '']
    b.move(3)
    assert b.board == ['', '', 'X', 'O', '', '', '']
    b.undo()
    ass(b.board, ['', '', 'X', '', '', '', ''])
    b.undo()
    assert b.board == ['', '', '', '', '', '', '']
```

```python
def test_check():
    ##### check_column() #####
    ass(Board('06060').check_column() , False)
    assert Board('0606060').check_column() is True

    ##### check_row() #####
    assert Board('06152').check_row() is False
    assert Board('0615243').check_row() is True

    ##### check_diagonal(1) ##### /
    assert Board('1122333352').check_diagonal(1) is False
    assert Board('112233335260').check_diagonal(1) is True

    ##### check_diagonal(-1) ##### \
    assert Board('112233335260').check_diagonal(-1) is False
    assert Board('000000111161446222633352').check_diagonal(-1) is True

    ##### check() #####
    assert Board('06060').check() is False
    assert Board('0606060').check() is True
    assert Board('0506060').check() is True
    assert Board('06060').check() is False
    assert Board('0616263').check() is True
    assert Board('0011224455663').check() is True
    assert Board('2030465').check() is True
    assert Board('20314').check() is False
    assert Board('65435234410111122120303').check() is True
    assert Board('65435234410111122120030').check() is False
    assert Board('203243665').check() is True
```

```python
def test_display_simple():
    b = Board()
    ass(b.display_simple().replace('\n', 'z') , 'z:z:z:z:z:z:z:z 012345z')
    ass(b.display_simple(), '''
:
:
:
:
:
:
:
 012345
''')
    b.move(2)
    ass(b.display_simple(), '''
:
:
:X
:
:
:
:
 012345
''')
    b.move(3)
    ass(b.display_simple(), '''
:
:
:X
:O
:
:
:
 012345
''')
    b.undo()
    ass(b.display_simple(), '''
:
:
:X
:
:
:
:
 012345
''')
    b.undo()
    ass(b.display_simple(), '''
:
:
:
:
:
:
:
 012345
''')
```

```python
def test_display():
    b = Board()
    disp = 'z5 . . . . . . .z4 . . . . . . .z3 . . . . . . .z2 . . . . . . .z1 . . . . . . .z0 . . . . . . .z  0 1 2 3 4 5 6z'
    ass(b.display().replace('\n', 'z') , disp)
    ass(b.display() , '''
5 . . . . . . .
4 . . . . . . .
3 . . . . . . .
2 . . . . . . .
1 . . . . . . .
0 . . . . . . .
  0 1 2 3 4 5 6
''')
    b.move(2)
    assert b.display() == '''
5 . . . . . . .
4 . . . . . . .
3 . . . . . . .
2 . . . . . . .
1 . . . . . . .
0 . . X . . . .
  0 1 2 3 4 5 6
'''
    b.move(3)
    assert b.display() == '''
5 . . . . . . .
4 . . . . . . .
3 . . . . . . .
2 . . . . . . .
1 . . . . . . .
0 . . X O . . .
  0 1 2 3 4 5 6
'''
    b.undo()
    assert b.display() == '''
5 . . . . . . .
4 . . . . . . .
3 . . . . . . .
2 . . . . . . .
1 . . . . . . .
0 . . X . . . .
  0 1 2 3 4 5 6
'''
    b.undo()
    assert b.display() == '''
5 . . . . . . .
4 . . . . . . .
3 . . . . . . .
2 . . . . . . .
1 . . . . . . .
0 . . . . . . .
  0 1 2 3 4 5 6
'''
```
